###Problema 3:
 La resta de un númeroEnunciado: Un número al que se le resta siete da como resultado 18.Paso 1: Planteamiento y resoluciónDefinir la variable: Sea $x$ el número desconocido.Plantear la ecuación: "Un número ($x$) menos ($-$) 7 es igual ($=$) a 18".$$x - 7 = 18$$Despejar la variable: El 7 que está restando pasa al otro lado sumando para dejar sola a la $x$.$$x = 18 + 7$$Resultado:$$x = 25$$

```python
# Problema 3: Hallar un número x tal que x - 7 = 18
# Se solicita al usuario ingresar los valores de la ecuación
resultado_total = int(input("Ingrese el resultado de la operación : "))
numero_restado = int(input("Ingrese el número que se está restando : "))

x = resultado_total + numero_restado

print(f"El número buscado es: {x}")
```

#### Problema 4:
 El triple de un númeroEnunciado: El triple de un número más cuatro es igual a 31.Paso 1: Resolución matemáticaVariable: Sea $x$ el número.Ecuación: "El triple de un número ($3x$) más ($+$) cuatro ($4$) es igual ($=$) a $31$".$$3x + 4 = 31$$Despeje (paso 1): El 4 pasa restando:$$3x = 31 - 4$$$$3x = 27$$Despeje (paso 2): El 3 que multiplica a la $x$ pasa dividiendo:$$x = \frac{27}{3}$$Resultado: $x = 9$


```python
# Problema 4: Hallar un número x tal que 3x + 4 = 31
# Definimos las entradas para las constantes de la ecuación
resultado_final = int(input("Ingrese el resultado de la igualdad : "))
numero_sumado = int(input("Ingrese el número que suma : "))
factor_multiplo = int(input("Ingrese el factor multiplicador : "))

paso_uno = resultado_final - numero_sumado

x = paso_uno / factor_multiplo

print(f"El número buscado es: {int(x)}")
```

#### Problema 5
Dimensiones de un terreno rectangularEnunciado: El largo mide 5m más que el ancho. Perímetro total = 50m.Paso 1: Resolución matemáticaVariables:Ancho: $x$Largo: $x + 5$Ecuación del perímetro: El perímetro es la suma de los 4 lados ($2 \cdot \text{largo} + 2 \cdot \text{ancho}$):$$2(x + 5) + 2(x) = 50$$Simplificación:$$2x + 10 + 2x = 50$$$$4x + 10 = 50$$Despeje:$$4x = 50 - 10$$$$4x = 40$$$$x = \frac{40}{4} = 10$$Resultados:Ancho: 10 metros.Largo: $10 + 5 =$ 15 metros.

```python
# Problema 5: Cálculo de dimensiones de un terreno (Perímetro = 50)
# Solicitamos los datos del problema geométrico
perimetro_total = int(input("Ingrese el perímetro total del terreno : "))
diferencia_largo = int(input("Ingrese la diferencia entre largo y ancho : "))

# La fórmula del perímetro 2(largo) + 2(ancho) = P se reduce a:
# 4*ancho + 2*diferencia = Perímetro
# Por lo tanto: ancho = (Perímetro - 2*diferencia) / 4

ancho = (perimetro_total - (2 * diferencia_largo)) / 4

largo = ancho + diferencia_largo

print(f"La medida del ancho es: {ancho} metros")
print(f"La medida del largo es: {largo} metros")
```
